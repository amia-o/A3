```python
import numpy 

import csv
#0 static
staticFile = open("static.csv")
staticReader = csv.reader(staticFile)
staticData = list(staticReader)

#1 driving
drivingFile = open("driving.csv")
drivingReader = csv.reader(drivingFile)
drivingData = list(drivingReader)

#2 walking
walkingFile = open("walking.csv")
walkingReader = csv.reader(walkingFile)
walkingData = list(walkingReader)

#3 running
runningFile = open("running.csv")
runningReader = csv.reader(runningFile)
runningData = list(runningReader)

#4 upstairs
upstairsFile = open("upstairs.csv")
upstairsReader = csv.reader(upstairsFile)
upstairsData = list(upstairsReader)

#initialize training list
X = []
Y = []

#initialize test list
Xt = []
Yt = []
```
